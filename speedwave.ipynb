```python
import torch
import torchaudio
import pydub
import glob
import os
import wave
folder_path = "E:/ASR/Rashin_Dataset_Total/Rashin_Dataset_Total/ASR12-DATA/mamo/mamo_voice"
wav_files = [f for f in os.listdir(folder_path) if f.endswith('.wav')]
for wav_file in wav_files:
    with wave.open(os.path.join(folder_path, wav_file), 'rb') as wf:
        print("File name:", wav_file)
        address=folder_path+"/"+wav_file
        new_folder_add="E:/ASR/Rashin_Dataset_Total/Rashin_Dataset_Total/ASR12-DATA/speedup12/mamo/mamo_voice/"
        finaladd=new_folder_add+wav_file
        audio = pydub.AudioSegment.from_file(address, format="wav")
        sample_rate = audio.frame_rate
        channels = audio.channels
        audio_data = torch.tensor(audio.get_array_of_samples(), dtype=torch.float32)
        transform = torchaudio.transforms.Resample(sample_rate, 22050)
        audio_data_resampled = transform(audio_data)
        audio_data_resampled_int = audio_data_resampled.numpy().astype("int16")
        audio_resampled = pydub.AudioSegment(audio_data_resampled_int.tobytes(),
                                            frame_rate=22050,
                                            sample_width=audio_data_resampled_int.dtype.itemsize,
                                            channels=channels)
        audio_resampled.export(address, format="wav")
        audio_faster = audio.speedup(playback_speed=1.4)
        audio_faster.export(finaladd, format="wav")
```
